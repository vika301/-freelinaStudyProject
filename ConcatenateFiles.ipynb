```python
import numpy as np
import pandas as pd
import glob
import os


#add path to your stored files; here its the current directory
path = os.getcwd()

#give path and name result file
savepath=os.getcwd(r'Concatinated.csv')
#concatinate your files
all_files = glob.glob(os.path.join(path, "*.csv"))
df = pd.concat((pd.read_csv(f) for f in all_files), ignore_index=True)
display(df)
#count duplicates
dup= df.duplicated()
dup.value_counts()
#save condatinated dataset in new path
df.to_csv(savepath, index=False, header=True)
```
